```python
import pandas as pd
import numpy as np
from faker import Faker
import  faker_commerce
from googletrans import Translator
```

`clientes`

```python


# Inicializar o Faker
fake = Faker('pt_BR')
fake.set_
# Gerar dados fictícios para clientes
clientes = []
for _ in range(30):
    cliente = {
        'ClienteID': fake.unique.random_number(digits=5),
        'Nome': fake.name(),
        'Email': fake.email(),
        'Telefone': fake.phone_number(),
        'Endereço': fake.address(),
        'Cidade': fake.city(),
        'Estado': fake.state(),
        'Prime' : fake.boolean(44),
        'Data de Nascimento' : fake.date_of_birth(minimum_age=18, maximum_age=65)
    }
    clientes.append(cliente)

# Criar DataFrame
clientes = pd.DataFrame(clientes)
clientes.head(3)
```

`produto`

```python
produtos = []
translator = Translator()

fake.add_provider(faker_commerce.Provider)
prod = [
    "TechBoost",
    "VitaFresh",
    "OndaLimpa",
    "GuardaInteligente",
    "FitPro",
    "EcoCarga",
    "RastroDeEnergia",
    "BioAtivo",
    "FluxoZen",
    "FrescuraPura",
    "EscudoUltra",
    "PeleRadiante",
    "SaborPuro",
    "SomQuântico",
    "MovimentoFlexível",
    "CasaSegura",
    "BrisaNatural",
    "RedeRápida",
    "CombustívelVerde",
    "OndaSolar",
    "CuidadoComÁgua",
    "VisãoMáxima",
    "CaminhoBrilhante"
]

for e, i in enumerate(range(23)):
    ecomerce = fake.ecommerce_name()

    produto = {
        'ProdutoID': fake.unique.random_number(digits=5),
        'Nome': prod[e],
        'Categoria': fake.random_element(elements=('Categoria A', 'A_', 'b','C','Aaa','P', 'C', 'CATEGORIA LETRA D', 'E')),
        'Preço': fake.random_int(min=10, max=1000),
    }
    produtos.append(produto)

produtos = pd.DataFrame(produtos)
produtos.head(3)
```

```python
preco = []
for i in produtos['Preço'].astype(str):
    r_ = np.random.normal(size=1)[0]
    if r_ > 0.3:
        i = 'R$ ' + i
    elif r_ > -0.1 and r_ <0.3:
        i = i +',00'
    else:
        i = i + ',00 reais'
    preco.append(i)
produtos['Preço'] = preco
produtos.head(3)
```

`vendas`

```python
from collections import OrderedDict

desconto = OrderedDict([
    ('0.10', 0.15),
    ('10%', 0.15),
    ('20%', 0.01),
    ('Nenhum', 0.45),
    ('0.1', 0.01),
    (False, 0.4),
    ('0.34%%', 0.02),
    ('0', 0.03),
    (0.0, 0.1),
    ('NÃO', 0.2),
])
vendas = []

for _ in range(1000):
    b = fake.boolean(66)
    if b == True:
        entrega = fake.date_between(start_date='today', end_date='+1m')
    else:
        entrega = np.nan
    venda = {
        'VendaID': fake.unique.random_number(digits=6),
        'ClienteID': fake.random_element(elements=clientes['ClienteID']),
        'ProdutoID': fake.random_element(elements=produtos['ProdutoID']),
        'Data Aprovação': fake.date_between(start_date='-1y', end_date='today'),
        'Data Entrega': entrega,
        'Forma Pagamento': fake.random_element(elements=('Cartão de Crédito', 'Boleto Bancário', 'No pix', 'Pix', 'Credito', 'Debito', 'Via Pix', 'Via Cartão')),
        'Desconto' : fake.random_element(elements=desconto),
    }
    vendas.append(venda)

vendas = pd.DataFrame(vendas)
vendas

```

```python
marcas = [
    'Chevrolet', 'Ford', 'Fiat', 'Volkswagen', 'Toyota',
    'Honda', 'Hyundai', 'Nissan', 'Renault', 'Jeep'
        ]

        # Lista de modelos por marca
modelos = {
    'Chevrolet': ['Onix', 'Cruze', 'Tracker'],
    'Ford': ['Ka', 'Fiesta', 'Ranger'],
    'Fiat': ['Uno', 'Mobi', 'Toro'],
    'Volkswagen': ['Gol', 'Polo', 'Amarok'],
    'Toyota': ['Corolla', 'Hilux', 'Etios'],
    'Honda': ['Civic', 'HR-V', 'CR-V'],
    'Hyundai': ['HB20', 'Creta', 'Tucson'],
    'Nissan': ['Versa', 'Kicks', 'March'],
    'Renault': ['Kwid', 'Sandero', 'Logan'],
    'Jeep': ['Renegade', 'Compass', 'Commander']
}
```

```python
veiculos = {"veiculosID":[], 'Nome':[]}
for i in marcas:
    veiculos['veiculosID'].append(fake.unique.random_number(digits=3))
    veiculos['Nome'].append(i)

veiculos = pd.DataFrame(veiculos)
veiculos
```

```python
veiculos.to_csv('veiculos.csv', index=False, encoding='utf-8')
```

```python
vendas_df = {
    'veiculoID':[],
    'Modelo':[],
    'Vendas':[],
    'Preço' :[],
    'data_vendas':[]
}

for i in range(100):
    data = fake.date_between(start_date='-3y', end_date='-1y')
    for i in range(50):
        marca  = fake.random_element(elements=marcas)
        veiculosID = veiculos[veiculos['Nome'] == marca]['veiculosID'].values[0]

        modelo = fake.random_element(elements=modelos[marca])


        if marca == 'Chevrolet':
            num_vendas = np.random.randint(30, 100)  # Changed variable name to num_vendas
            preco = 'R$ '+str(np.random.randint(10000, 20000))
        elif marca == 'Ford':
            num_vendas = np.random.randint(1, 10)  # Changed variable name to num_vendas
            preco = 'R$'+str(np.random.randint(20000, 30000))
        elif marca == 'Fiat':
            num_vendas = np.random.randint(40, 120)  # Changed variable name to num_vendas
            preco = 'R$ '+str(np.random.randint(20000, 35000))

        elif marca == 'Nissan' or marca == 'Toyota' or marca == 'Hyundai':
            num_vendas = np.random.randint(1, 20)  # Changed variable name to num_vendas
            preco  = 'R$ '+str(np.random.randint(80000, 100000))
        else:
            num_vendas = np.random.randint(1, 5)  # Changed variable name to num_vendas
            preco = 'R$ '+str(np.random.randint(45000, 70000))


        vendas_df['veiculoID'].append(veiculosID) # Appending to the dictionary vendas_df
        vendas_df['Vendas'].append(num_vendas) # Using num_vendas here
        vendas_df['Preço'].append(preco)
        vendas_df['Modelo'].append(modelo)
        vendas_df['data_vendas'].append(data)


vendas_df = pd.DataFrame(vendas_df) # Creating the DataFrame vendas_df
vendas_df
```

```python
vendas_df.to_csv('vendas.csv', index=False, encoding='utf-8')
```
